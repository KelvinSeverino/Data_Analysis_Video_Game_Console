```python
#Importando todas as Bibliotecas necessárias para realização da análise 
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
#Leitura de Arquivo
videogames = pd.read_csv('C:\\Users\\Kel\Desktop\\Plataformas_Consoles.csv')
```

```python
#Exibindo todas linhas da Base de Dados
videogames.head(34)
```

```python
#Tipo de dado em cada Coluna
videogames.dtypes
```

```python
#Quantidade de Linhas e colunas na Base de Dados
videogames.shape
```

```python
#Renomeando colunas
videogames.columns = ["Ranking", "Plataforma", "Fabricante", "Ano de Lançamento", "América do Norte",
                     "Europa", "Japão", "Resto do Mundo", "Vendas Globais",
                     "Games Lançados"]
```

```python
#Exibindo todas Linhas da Base de Dados com as devidas alterações
videogames.head(34)
```

```python
#Retornando valores Nulos
videogames[videogames["Vendas Globais"].isnull()].head()
```

```python
#Exibindo a quantidade de consoles por Fabricante
videogames["Fabricante"].value_counts()
```

```python
#Gráfico da quantidade de consoles por fabricante
videogames["Fabricante"].value_counts().plot()
plt.show()
```

```python
#Criando um gráfico apenas uma linha de código
videogames["Fabricante"].value_counts().head(10).plot(kind="bar", figsize=(15,5), grid = False, rot = 0, color = "blue")

#Acrescentando detalhes ao gráfico
plt.title("As 10 Fabricantes com mais consoles lançados")
plt.xlabel("Fabricantes")
plt.ylabel("Quantidade de Plataformas/Consoles")
```

```python
#Criando um gráfico apenas uma linha de código
top_10_anos = videogames[["Plataforma", "Vendas Globais"]].head(5).set_index("Plataforma").sort_values("Vendas Globais", ascending=True)
top_10_anos.plot(kind="barh", figsize=(15,7), grid = False, color="darkred", legend=False)

#Acrescentando detalhes ao gráfico
plt.title("Os 5 Consoles mais vendidos do Mundo")
plt.xlabel("Total de Vendas em milhões de Unidades")
plt.show()
```

```python
#Criando um gráfico apenas uma linha de código
top_10_anos = videogames[["Plataforma", "América do Norte"]].head(5).set_index("Plataforma").sort_values("América do Norte", ascending=True)
top_10_anos.plot(kind="barh", figsize=(12,7), grid = False, color="red", legend=False)

#Acrescentando detalhes ao gráfico
plt.title("Os 5 Consoles mais vendidos na América do Norte")
plt.xlabel("Total de Vendas em milhões de Unidades")
plt.show()
```

```python
#Criando um gráfico apenas uma linha de código
top_10_anos = videogames[["Plataforma", "Resto do Mundo"]].head(5).set_index("Plataforma").sort_values("Resto do Mundo", ascending=True)
top_10_anos.plot(kind="barh", figsize=(11,7), grid = False, color="darkgreen", legend=False)

#Acrescentando detalhes ao gráfico
plt.title("Os 5 Consoles mais vendidos no Resto do Mundo")
plt.xlabel("Total de Vendas em milhões de Unidades")
plt.show()
```

```python
#Criando um gráfico apenas uma linha de código
top_10_anos = videogames[["Plataforma", "Games Lançados"]].head(5).set_index("Plataforma").sort_values("Games Lançados", ascending=True)
top_10_anos.plot(kind="barh", figsize=(11,7), grid = False, color="grey", legend=False)

#Acrescentando detalhes ao gráfico
plt.title("Os 7 Consoles com mais Games Lançados de Todos os Tempos")
plt.xlabel("Total de Games Lançados em Mil Unidades")
plt.show()
```
